```python
# Task 01: Visualizing Age and Gender Distribution

# Step 1: Importing required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Just making the plots look cleaner
sns.set(style="whitegrid")
# Step 2: Loading the dataset
# Using Titanic dataset because it has both Age and Gender info

data_url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv"
df = pd.read_csv(data_url)

# Let’s take a quick look at the first few rows
df[['Age', 'Sex']].head()
# Step 3: Checking data types and missing values
print("Basic Info:")
df[['Age', 'Sex']].info()

# Check if Age or Gender have any missing values
print("\nMissing values:\n")
print(df[['Age', 'Sex']].isnull().sum())
# Step 4: Bar Chart – Gender Distribution

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex', palette='pastel')

plt.title('Gender Distribution on the Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
# Step 5: Histogram – Age Distribution
# Dropping missing ages just to be safe

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', bins=15, kde=True, color='skyblue')

plt.title('Age Distribution on the Titanic')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.tight_layout()
plt.show()
```

### Task 01 Summary

In this task, I used the Titanic dataset because it includes `Age` and `Sex` columns, which are perfect for visualizing both continuous and categorical variables.

**Gender Distribution:**  
From the bar chart, we can see that there were more males than females on board.

**Age Distribution:**  
Most passengers were in the 20–40 age range. There’s a wider spread toward older ages too, but fewer kids and teens.

This kind of quick EDA helps get a better understanding of the dataset before moving on to modeling or deeper analysis.
